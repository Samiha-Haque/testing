<b>Comparing the cotinine and hydroxycotinine levels of smokers and non-smokers</b>

According to metadat, Cotinine and Hydroxycotinine levels indicate active smoking and 0.015 is the threshold value, i.e. any person with levels over 0.015  is considered to have used tobacco.

I created a function called map_smoking which takes the row of the dataframe and then check the Cotinine and Hydroxycotinine levels, if both the values are less than 0.015(threshold), then 'Non smoker' is returned. Then I used only Cotinine levels(the main metabolite of nicotine, is typically considered the most reliable marker for nicotine exposure.) to further divide into 'Light smoker', 'Moderate smoker', 'Heavy smoker' and 'Extreme level smoker'. Finally, I used apply function of the dataframe to create a new column 'Smoking level' to map the smoking levels to individual persons.

```python
def map_smoking(row):
    cotinine_level=row['Cotinine']
    hydroxycotinine_level=row['Hydroxycotinine']
    if cotinine_level<0.015 and hydroxycotinine_level<0.015:
        return 'Non smoker'
    elif cotinine_level<=1:
        return 'Light smoker'
    elif cotinine_level<=10:
        return 'Moderate smoker'
    elif cotinine_level<=100:
        return 'Heavy smoker'
    elif cotinine_level>100:
        return 'Extreme level smoker'

x['Smoking_level'] = x.apply(map_smoking,axis=1)
x.head()
```

I filtered the dataset using 'Smoking_level' and selected only rows which had 'Smoking_level' as 'Non smoker' and stored the subset in non_smokers variable. Next I again filtered the dataset to selected all the rows where 'Smoking_level' is not 'Non smoker' using ~ and stored the subset in smokers variable.

```python
non_smokers=x.loc[x['Smoking_level']=='Non smoker'].reset_index(drop=True)
smokers=x.loc[~(x['Smoking_level']=='Non smoker')].reset_index(drop=True)
smokers.rename_axis('index',inplace=True)
```

#### Reflection on NHANES data privacy rules

During NHANES surveys, the participants' names remain disconnected from their questionnaire responses and test outcomes. All information that could potentially identify individuals or their families, such as names, addresses, phone numbers, and workplaces, is eliminated. Additionally, data concerning very specific geographical areas, which could potentially lead to the identification of individuals, are not disclosed. Each participant receives a distinct identification number solely for the survey. Participation is voluntary, and all responses are kept confidential. Besides that, NHANES databases are safeguarded by passwords and encryption, accessible only to authorized personnel. NCHS integrates participants' protected data with data from numerous other survey participants. The combined data are carefully checked to prevent any linkage to individual participants, families, or persons.

NHANES places a high emphasis on data privacy and ethics and therefore, I made sure that my analysis is conducted with integrity and respects the rights and confidentiality of participants.
